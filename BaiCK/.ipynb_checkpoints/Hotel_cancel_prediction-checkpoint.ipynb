# Dự đoán hủy phòng dùng thuật toán Random Forest
![bookinghotel](Nhom21.jpg)
### Về các cột trong Dataset
 * Booking_ID: mã duy nhất của mỗi lượt đặt
* no_of_adults: số lượn người lớn
* no_of_children: số lượng trẻ em
* no_of_weekend_nights: Số đêm cuối tuần (thứ bảy hoặc chủ nhật), khách đã ở lại hoặc đặt chỗ ở khách sạn
* no_of_week_nights: Số đêm trong tuần (Thứ Hai đến Thứ Sáu), khách đã ở lại hoặc đặt chỗ ở khách sạn
* type_of_meal_plan: Loại kế hoạch bữa ăn được đặt bởi khách hàng:
* required_car_parking_space: Khách hàng có yêu cầu một chỗ đậu xe không? (0 - No, 1- Yes)
* room_type_reserved: Loại phòng được đặt bởi khách hàng. Các giá trị được mã hóa bằng INN Hotels.
* lead_time: Số ngày giữa ngày đặt phòng và ngày đến
* arrival_year: Năm đến
* arrival_month: Tháng đến
* arrival_date: Ngày của tháng
* market_segment_type: Chỉ định phân khúc thị trường.
* repeated_guest: Khách hàng có phải là khách lặp đi lặp lại không? (0 - No, 1- Yes)
* no_of_previous_cancellations: Số lượng đặt phòng trước đó đã bị khách hàng hủy bỏ trước khi đặt phòng hiện tại
* no_of_previous_bookings_not_canceled: Số lượng đặt phòng trước đó không bị khách hàng hủy bỏ trước khi đặt phòng hiện tại
* avg_price_per_room: giá trung bình mỗi ngày đặt phòng; Giá của các phòng là năng động. (tính bằng euro)
* no_of_special_requests: Tổng số yêu cầu đặc biệt do khách hàng thực hiện (ví dụ: tầng cao, view xem từ phòng, v.v.)
* booking_status: Cờ cho biết việc đặt phòng có bị hủy hay không

==>Link Dataset: https://www.kaggle.com/datasets/ahsan81/hotel-reservations-classification-dataset

## Tiền xử lí dữ liệu

```python
df= df.drop('Booking_ID', axis = 1)
```

```python
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object': 
        df[col] = le.fit_transform(df[col].astype(str))
```

## Huấn luyện mô hình với Random Forest

```python
# Chia thành input và output
X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print("X_train: ",X_train.shape)
print("Y_train: ",Y_train.shape)
print("X_test: ",X_test.shape)
print("Y_test: ",Y_test.shape)
```

```python
# Huấn luyện mô hình Random Forests
model = RandomForestClassifier(n_estimators=100, max_features='sqrt')
model.fit(X_train, Y_train)

# Dự đoán kết quả trên tập kiểm tra
Y_pred = model.predict(X_test)

# Đánh giá
accuracy = accuracy_score(Y_test, Y_pred)
print("Độ chính xác(accuracy):", accuracy)

report = classification_report(Y_test, Y_pred)
print("Đánh giá mô hình:")
print(report)
```
